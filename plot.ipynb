```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use(['seaborn-colorblind', 'seaborn-talk'])
plt.rcParams['figure.dpi'] = 300
plt.rcParams['savefig.dpi'] = 300
plt.rcParams['savefig.bbox'] = 'tight'
plt.rcParams['font.sans-serif'] = ['Taipei Sans TC Beta'] # to plot Chinese words properly

import datetime
date = datetime.datetime.now().strftime('%Y%m%d')
```

```python
no_df = pd.DataFrame({
    'year': range(2000, 2020), 
    'NO3': np.random.randint(10, 100, 20)})

plt.figure(figsize=(5, 3.7))
plt.plot(no_df.year, no_df.NO3, alpha=.5, c='gray')
mask = no_df.NO3 < 25 
plt.scatter(no_df.year[~mask], no_df.NO3[~mask], label='未符合', edgecolors='gray', facecolors='none')
plt.scatter(no_df.year[mask], no_df.NO3[mask], label='符合', c='black')
plt.xticks(range(2000, 2021, 5))
plt.xlabel('年份')
plt.ylabel('硝酸鹽氮 (mg/L)')
plt.legend()
plt.savefig('results/example_plot.png')
```
